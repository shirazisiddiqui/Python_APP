```python
import tkinter as tk
from tkinter import *
import requests
import time
from PIL import ImageTk, Image
from tkinter import messagebox
```

```python
## Create the main window
## Weather function

def getWeather(canvas):
    city = textfield.get()
    # Add your API key here
    apikey = "BEREKZT5NRCTEL4AT3JD5DRP4"
    url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/" + city + "?key=" + apikey 
    result = requests.get(url)
    if result:
        ###Extract current weather info from JSON
        json_data = result.json()
        cond = json_data['days'][0]['hours'][0]
        temp = cond['temp']
        feelslike = cond['feelslike']
        humidity = cond['humidity']
        pressure = cond['pressure']
        conditions = cond['conditions']
        visibility = cond['visibility']
        icon = cond['icon']
        
        #Extract DAY+1 weather info from JSON
        cond1 = json_data['days'][1]
        Date_1 = cond1['datetime']
        forecast_day1 = cond1['conditions']
        Temp1_min = cond1['tempmin']
        Temp1_max = cond1['tempmax']
        icon1 = cond1['icon']
    
        #Extract DAY+2 weather info from JSON
        cond2 = json_data['days'][2]
        Date_2 = cond2['datetime']
        forecast_day2 = cond2['conditions']
        Temp2_min = cond2['tempmin']
        Temp2_max = cond2['tempmax']
        icon2 = cond2['icon']
    
        #Extract DAY+3 weather info from JSON
        cond3 = json_data['days'][3]
        Date_3 = cond3['datetime']
        forecast_day3 = cond3['conditions']
        Temp3_min = cond3['tempmin']
        Temp3_max = cond3['tempmax']
        icon3 = cond3['icon']
    
        #Extract DAY+4 weather info from JSON
        cond4 = json_data['days'][4]
        Date_4 = cond4['datetime']
        forecast_day4 = cond4['conditions']
        Temp4_min = cond4['tempmin']
        Temp4_max = cond4['tempmax']
        icon4 = cond4['icon']
    
        #Extract DAY+5 weather info from JSON
        cond5 = json_data['days'][5]
        Date_5 = cond5['datetime']
        forecast_day5 = cond5['conditions']
        Temp5_min = cond5['tempmin']
        Temp5_max = cond5['tempmax']
        icon5 = cond5['icon']
        
        image['bitmap'] = '/Users/shirazisiddiqui/Downloads/weather_icons/{}.png'.format(icon)
        image_1['bitmap'] = '/Users/shirazisiddiqui/Downloads/weather_icons/{}.png'.format(icon1)
        image_2['bitmap'] = '/Users/shirazisiddiqui/Downloads/weather_icons/{}.png'.format(icon2)
        image_3['bitmap'] = '/Users/shirazisiddiqui/Downloads/weather_icons/{}.png'.format(icon3)
        image_4['bitmap'] = '/Users/shirazisiddiqui/Downloads/weather_icons/{}.png'.format(icon4)
        image_5['bitmap'] = '/Users/shirazisiddiqui/Downloads/weather_icons/{}.png'.format(icon5)
        
        
        final_info = "Current Temperature of "+city+" : "+str(temp)+"°F"+"\n"+"Feelslike : "+str(feelslike)+"°F"+'\n'+"Weather Condition : "+str(conditions)+'\n'+"Humidity is : "+str(humidity)+'\n'+"Pressure is : "+str(pressure)
        city_label1.config(text = final_info)
        image_1_data = "Five days weather forecast"+'\n'+'\n'+ "On "+ str(Date_1)+","+"Weather forecast is "+str(forecast_day1)+","+"MaxTemp is "+str(Temp1_max)+"°F"+" and "+"MinTemp "+str(Temp1_min)+"°F"+' '
        image_2_data = "On "+ str(Date_2)+","+"Weather forecast is "+str(forecast_day2)+","+"MaxTemp is "+str(Temp2_max)+"°F"+" and "+"MinTemp is "+str(Temp2_min)+"°F"+' '
        image_3_data = "On "+ str(Date_3)+","+"Weather forecast is "+str(forecast_day3)+","+"MaxTemp is "+str(Temp3_max)+"°F"+" and "+"MinTemp is "+str(Temp3_min)+"°F"+' '
        image_4_data = "On "+ str(Date_4)+","+"Weather forecast is "+str(forecast_day4)+","+"MaxTemp is "+str(Temp4_max)+"°F"+" and "+"MinTemp is "+str(Temp4_min)+"°F"+' '
        image_5_data = "On "+ str(Date_5)+","+"Weather forecast is "+str(forecast_day5)+","+"MaxTemp is "+str(Temp5_max)+"°F"+" and "+"MinTemp is "+str(Temp5_min)+"°F"+' '
        image_1.config(text = image_1_data)
        image_2.config(text = image_2_data)
        image_3.config(text = image_3_data)
        image_4.config(text = image_4_data)
        image_5.config(text = image_5_data)

              
        return final_info,image_1_data,image_2_data,image_3_data,image_4_data,image_5_data
    else:
        messagebox.showerror('Error' ,'ERROR DETECTED! CANNOT FIND CITY : {}'.format(city))

# Create window    
canvas = tk.Tk()
canvas.geometry("1500x1500+500+600")
canvas.resizable(width=False, height=False)
canvas.title("Weather Broadcast")
f = ("poppins",15,"bold")
t = ("poppins",20,"bold")


# Create and configure labels and entry fields
textfield = tk.Entry(canvas, justify = 'center',font= t)
textfield.pack(pady = 20)
textfield.focus()
textfield.bind('<Return>', getWeather)


# Add image icon for current weather
image = tk.Label(canvas , bitmap='', text = "Current Weather Condition",font = t,compound='top')
image.pack()

##Creation of label 1 and pack
city_label1 = tk.Label(canvas,font = t , text ='Welcome to the Weather Broadcast',compound='center')
city_label1.place(x = 0, y = 0,anchor=CENTER)
city_label1.pack(pady = 10)

canvas.rowconfigure(0, weight=1)
canvas.columnconfigure(0, weight=1)

# Add image icon for 5 days forecast weather
image_1 = tk.Label(canvas , bitmap='' ,font = f,compound='right')
image_1.pack()

image_2 = tk.Label(canvas , bitmap='',font = f,compound='right')
image_2.pack()

image_3 = tk.Label(canvas , bitmap='',font = f,compound='right')
image_3.pack()

image_4 = tk.Label(canvas , bitmap='',font = f,compound='right')
image_4.pack()

image_5 = tk.Label(canvas , bitmap='',font = f,compound='right')
image_5.pack()

# Start the GUI main loop
canvas.mainloop()
```
